```python
import cv2
import mediapipe as mp
import numpy as np
from mediapipe.tasks import python
from mediapipe.tasks.python import vision

# تجهيز الإعدادات
model_path = 'hand_landmarker.task' # اتأكدي إن الملف في نفس المكان
base_options = python.BaseOptions(model_asset_path=model_path)
options = vision.HandLandmarkerOptions(
    base_options=base_options,
    running_mode=vision.RunningMode.VIDEO,
    num_hands=1 # هنركز على إيد واحدة عشان الدقة
)
landmarker = vision.HandLandmarker.create_from_options(options)

print("Model Loaded Successfully!")
```

```python
import cv2
import mediapipe as mp
import time

cap = cv2.VideoCapture(0) # جربي 0 أو 1 حسب Iriun

# أرقام النقاط
finger_tips = [8, 12, 16, 20]
finger_pip_joints = [6, 10, 14, 18]

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break

    frame = cv2.flip(frame, 1)
    h, w, _ = frame.shape
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=image_rgb)

    # الحل السحري لمشكلة الـ Timestamp
    timestamp = int(time.time() * 1000)
    results = landmarker.detect_for_video(mp_image, timestamp)

    status = "Waiting..." 
    color = (200, 200, 200)

    if results.hand_landmarks:
        prev_status = "Waiting..."
counter = 0
        for hand_landmarks in results.hand_landmarks:
            opened_fingers = 0
            for i in range(4):
                # بنقارن مكان طرف الصباع بمفصله
                if hand_landmarks[finger_tips[i]].y < hand_landmarks[finger_pip_joints[i]].y:
                    opened_fingers += 1
                # Thumb
                if hand_landmarks[4].x < hand_landmarks[3].x:
                    opened_fingers += 1
            
            if opened_fingers <= 1: # لو صباع واحد أو صفر مرفوعين يبقى مقفولة
                status = "Hand Closed"
                color = (0, 0, 255)
            else:
                status = "Hand Open"
                color = (0, 255, 0)

            # رسم النقط
            for lm in hand_landmarks:
                cx, cy = int(lm.x * w), int(lm.y * h)
                cv2.circle(frame, (cx, cy), 3, color, -1)

    # تصميم الشاشة
    cv2.rectangle(frame, (0, 0), (300, 60), (0, 0, 0), -1)
    cv2.putText(frame, status, (10, 45), cv2.FONT_HERSHEY_SIMPLEX, 1.2, color, 3)

    cv2.imshow('Hand Status Detection', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
